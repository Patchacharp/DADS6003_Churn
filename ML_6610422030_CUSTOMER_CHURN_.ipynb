# <span style="font-family:serif; font-size:28px;">Data Preprocessing</span>
<a id = "datapreprocessing" ></a>

one hot encoding all the categorical features describing different statuses of a customer

```python
df.gender = df.gender.replace({'Female':0,'Male':1})
df.Partner = df.Partner.replace({'Yes':1,'No':0})
df.Dependents = df.Dependents.replace({'Yes':1,'No':0})
df.PhoneService = df.PhoneService.replace({'Yes':1,'No':0})
df.PaperlessBilling = df.PaperlessBilling.replace({'Yes':1,'No':0})
df.Churn = df.Churn.replace({'Yes':1,'No':0})
df.MultipleLines = df.MultipleLines.replace({'Yes':'Yes multi','No':'No multi'})
df.InternetService = df.InternetService.replace({'No':'No internet'})
df.OnlineSecurity = df.OnlineSecurity.replace({'Yes':'Yes Secur','No':'No Secur','No internet service':'No internet Secur'})
df.OnlineBackup = df.OnlineBackup.replace({'Yes':'Yes Back','No':'No Back','No internet service':'No internet Back'})
df.DeviceProtection = df.DeviceProtection.replace({'Yes':'Yes Dev','No':'No Dev','No internet service':'No internet Dev'})
df.TechSupport = df.TechSupport.replace({'Yes':'Yes Tec','No':'No Tec','No internet service':'No internet Tec'})
df.StreamingTV = df.StreamingTV.replace({'Yes':'Yes TV','No':'No TV','No internet service':'No internet TV'})
df.StreamingMovies = df.StreamingMovies.replace({'Yes':'Yes Mov','No':'No Mov','No internet service':'No internet Mov'})
```

```python
df = pd.concat([df,pd.get_dummies(df['MultipleLines'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['InternetService'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['OnlineSecurity'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['OnlineBackup'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['DeviceProtection'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['TechSupport'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['StreamingTV'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['StreamingMovies'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['Contract'])],axis=1)
df = pd.concat([df,pd.get_dummies(df['PaymentMethod'])],axis=1)
df['churn'] = df['Churn']
df.drop(columns = ['Churn','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaymentMethod'],inplace=True)
df = df.astype(int)
df
```

#Data Upsampling Using SMOTE

```python
oversample = SMOTE()
X, y = oversample.fit_resample(df[df.columns[:-1]], df[df.columns[-1]])
usampled_df = X.assign(Churn = y)
usampled_df
```

```python
X = usampled_df.drop(columns = ['Churn'])
y = usampled_df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 40, stratify=y)
```

# <span style="font-family:serif; font-size:28px;">Machine Learning Model Evaluations and Predictions</span>

**These are 5 models that can improve the weighted avg f1-score of the original notebook**

<a id = "103" ></a>
#### <b> Random Forest</b>
<a id = "rf" ></a>

```python
model_rf = RandomForestClassifier(n_estimators=500 , oob_score = True, n_jobs = -1,
                                  random_state =50, max_features = "auto",
                                  max_leaf_nodes = 30)
model_rf.fit(X_train, y_train)

# Make predictions
prediction_test = model_rf.predict(X_test)
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
print(classification_report(y_test, prediction_test))
```

<a id = "104" ></a>
#### <b>Logistic Regression</b>
<a id = "lr" ></a>

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
accuracy_lr = lr_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy_lr)
```

```python
lr_pred= lr_model.predict(X_test)
report = classification_report(y_test,lr_pred)
print(report)
```

<a id = "106" ></a>
#### **AdaBoost Classifier**
<a id = "ada" ></a>

```python
a_model = AdaBoostClassifier()
a_model.fit(X_train,y_train)
a_preds = a_model.predict(X_test)
print("AdaBoost Classifier accuracy")
metrics.accuracy_score(y_test, a_preds)
```

```python
print(classification_report(y_test, a_preds))
```

<a id = "107" ></a>
#### **Gradient Boosting Classifier**
<a id = "gb" ></a>

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
print("Gradient Boosting Classifier", accuracy_score(y_test, gb_pred))
```

```python
print(classification_report(y_test, gb_pred))
```

<a id = "108" ></a>
#### **Voting Classifier**
<a id = "vc" ></a>

```python
from sklearn.ensemble import VotingClassifier
clf1 = GradientBoostingClassifier()
clf2 = LogisticRegression()
clf3 = AdaBoostClassifier()
eclf1 = VotingClassifier(estimators=[('gbc', clf1), ('lr', clf2), ('abc', clf3)], voting='soft')
eclf1.fit(X_train, y_train)
predictions = eclf1.predict(X_test)
print("Final Accuracy Score ")
print(accuracy_score(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```
